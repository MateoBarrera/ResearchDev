```python
# Total alternatives 56
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

# Read the Excel file into a pandas DataFrame
df = pd.read_excel("../lista_de_resultados_CI_56a_16-07-2024.xlsx")
df.rename(columns={"Unnamed: 0":"Alternative"}, inplace=True)
df.head()
```

```python
df_dummy = df.copy()
df_dummy.drop(columns=["Alternative"], inplace=True)
# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(df_dummy.corr(), annot=True)

# Display the heatmap
plt.show()
```

```python
# Read the Excel file into a pandas DataFrame
df_categorical = pd.read_excel("../lista_de_resultados_56a_16-07-2024.xlsx")
df_categorical.rename(columns={"Unnamed: 0": "Alternative"}, inplace=True)
df_categorical.head()
```

```python
df_dummy_categorical = df_categorical.copy()
df_dummy_categorical.drop(columns=["Alternative"], inplace=True)
# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
#sns.heatmap(df_dummy_categorical.corr(method='kendall'), annot=True)
ax = sns.heatmap(df_dummy_categorical.corr(), annot=True)
ax.set(xlabel="", ylabel="")

# Display the heatmap
plt.show()
```

```python
df_environmental_ci = df_dummy[
    [
        "Scenario 16",
        "Scenario 17",
        "Scenario 18",
        "Scenario 19",
        "Scenario 20",
        "Scenario 21",
        "Scenario 22",
    ]
].copy()

df_environmental_cg = df_dummy_categorical[
    [
        "Scenario 16",
        "Scenario 17",
        "Scenario 18",
        "Scenario 19",
        "Scenario 20",
        "Scenario 21",
        "Scenario 22",
    ]
].copy()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle("Correlation Heatmap\n Environmental dimension")
sns.heatmap(df_environmental_ci.corr(), annot=True, ax=ax1)
sns.heatmap(
    df_environmental_cg.corr(), annot=True, ax=ax2
)
# Display the heatmap
plt.show()
```
