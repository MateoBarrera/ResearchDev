```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

# Read the Excel file into a pandas DataFrame
df = pd.read_excel("../lista_de_resultados_CI.xlsx")
df.rename(columns={"Unnamed: 0":"Alternative"}, inplace=True)
df.head()
```

```python
df_dummy = df.copy()
df_dummy.drop(columns=["Alternative"], inplace=True)
# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(df_dummy.corr(), annot=True, cmap='coolwarm')

# Display the heatmap
plt.show()
```

```python
# Read the Excel file into a pandas DataFrame
df_categorical = pd.read_excel("../lista_de_resultados.xlsx")
df_categorical.rename(columns={"Unnamed: 0": "Alternative"}, inplace=True)
df_categorical.head()
```

```python
df_dummy_categorical = df_categorical.copy()
df_dummy_categorical.drop(columns=["Alternative"], inplace=True)
# Create a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(df_dummy_categorical.corr(method='kendall'), annot=True, cmap="coolwarm")

# Display the heatmap
plt.show()
```
